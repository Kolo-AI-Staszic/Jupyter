Mnożenie przez sklalar

$$3*\mathbf{v}=3*\begin{bmatrix} 1 \\ 1 \\ 1 \end{bmatrix}=\begin{bmatrix} 3*1 \\ 3*1 \\ 3*1 \end{bmatrix}=\begin{bmatrix} 3 \\ 3 \\ 3 \end{bmatrix}$$

Dodawanie wektorów

$$\mathbf{v}_1 = \begin{bmatrix} 1 \\ 7  \end{bmatrix}, \mathbf{v}_2 = \begin{bmatrix} 4 \\ 2 \end{bmatrix}$$
$$\mathbf{v}_1 + \mathbf{v}_2 = \begin{bmatrix} 1 \\ 7  \end{bmatrix} + \mathbf{v}_2 = \begin{bmatrix} 4 \\ 2 \end{bmatrix} = \begin{bmatrix} 5 \\ 9 \end{bmatrix}$$

```python
plt.figure(figsize=(8,8))
wektor1=mpimg.imread('dodawanie.png')
imgplot = plt.imshow(wektor1)
plt.show()
```

Norma/długość wektora
$$\mathbf{v} = \begin{bmatrix} 1 \\ 7 \\ 8 \\ 3 \end{bmatrix}$$

$$||\mathbf{v}|| = \sqrt{1^2+7^2+8^2+3^2} = \sqrt{123}$$

Mnożenie wektorów (iloczyn skalarny)
$$\mathbf{v}_1 * \mathbf{v}_2 = \begin{bmatrix} 1 \\ 7  \end{bmatrix} * \begin{bmatrix} 4 \\ 2 \end{bmatrix} = (1*4+7*2) = 18 $$
$$lub$$
$$\mathbf{v}_1 * \mathbf{v}_2 = ||\mathbf{v}_1|| * ||\mathbf{v}_2|| * \cos \alpha, \cos \alpha = \frac{\mathbf{v}_1 * \mathbf{v}_2}{|\mathbf{v}_1| * |\mathbf{v}_2|}$$



Podobieństwo cosinusowe
$$\cos \alpha = \frac{\mathbf{v}_1 * \mathbf{v}_2}{|\mathbf{v}_1| * |\mathbf{v}_2|}$$

Przyjmujemy wartości od -1 (brak podobieństwa) do 1 ($\mathbf{v}_1$ jest takie samo jak $\mathbf{v}_2$)

Transponowanie wektorów
$$\mathbf{v} = \begin{bmatrix} 1 \\ 7 \\ 8 \\ 3 \end{bmatrix} = \begin{bmatrix} 1 && 7 && 8 && 3 \end{bmatrix}^T$$
